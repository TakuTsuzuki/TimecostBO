```python
%matplotlib inline
import matplotlib.pyplot as plt
import sys
import time
import numpy as np 
```

### ナイーブ再帰

```python
def culc_fib_rec(n):
    if n > 1:
        return culc_fib_rec(n-1)+culc_fib_rec(n-2)
    elif n == 1:
        return 1
    else:
        return 0
```

```python
time_ = []
for i in range(1,35):
    start = time.time()
    culc_fib_rec(i)
    end = time.time()
    t = end-start
    time_.append(t)
```

```python
plt.plot(np.log10(time_))
```

### メモ化

```python
fibonacci_list = {}
```

```python
def culc_fib_memo(n):
    global fibonacci_list
    if n == 1:
        return 1
    elif n == 0:
        return 0
    if not n in fibonacci_list:
        fibonacci_list[n] = culc_fib_memo(n-1) + culc_fib_memo(n-2)
    return fibonacci_list[n]
```

```python
time_ = []
for i in range(1,500):
    start = time.time()
    culc_fib_memo(i)
    end = time.time()
    t = end-start
    time_.append(t)
```

```python
plt.plot(time_)
```
